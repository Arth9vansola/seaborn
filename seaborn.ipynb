```python
df = sns.load_dataset("tips")
df.head()
```

# lineplot

```python
sns.lineplot(x="day", y="total_bill", data=df, hue='sex', style='time', palette="flare")
plt.show()
```

# barplot

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x = "sex", y = "fare", data=df, hue="who", color="yellow", palette="flare", saturation=1, ci=False)
plt.show()
```

```python
from numpy import mean
sns.barplot(x = "fare", y = "sex", data=df, hue="who", color="yellow", palette="flare", saturation=1, estimator=mean, facecolor=[1,1,1,0], edgecolor="0.2")
plt.show()
```

```python
from numpy import median
sns.barplot(x = "fare", y = "sex", data=df, hue="who", color="yellow", palette="flare", saturation=1, estimator=median)
plt.show()
```

```python
sns.barplot(x = "sex", y = "fare", data=df, color="yellow", order=["female", "male"])
plt.show()
```

# histogram

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.histplot(x='total_bill', data = df, discrete=False, hue="sex", kde = True)
plt.show() # kde = kernel density estimation
```

# scatterplot

```python
df = sns.load_dataset("exercise")
df.head()
```

```python
sns.scatterplot(x="time", y="pulse", data=df)
plt.show()
```

```python
sns.scatterplot(x="time", y="pulse", data=df, hue="kind", palette="flare", size="diet", marker='o', alpha=0.7) # alpha:it will showcase the transparency means how it show or not
plt.show()
```

# heatmap

```python
data = np.random.randint(low = 1, high = 100, size=(10,10))
print(data)
```

```python
sns.heatmap(data=data, cmap='flare', linecolor="white", linewidth="2")
plt.show()
```

```python
sns.heatmap(data=data, cmap='flare', annot= True, cbar = False)
plt.show()
```

```python
sns.heatmap(data=data, cmap='flare', annot= True, cbar = False, xticklabels = False , yticklabels = False)
plt.show()
```

# countplot: same as barplot but here we give just and x and it will automatically take value of y

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.countplot(x='sex',data=df)
plt.show()
```

```python
sns.countplot(x='sex',data=df, hue='smoker', saturation=0.9, alpha=0.9)
plt.show() 
# alpha : transparency of background overall canvas
# saturation : transparency of main content here is bars
```

```python
sns.countplot(x='sex',data=df, hue='smoker', palette='dark:red')
plt.show() 
```

```python
sns.countplot(x='sex',data=df, hue='smoker', palette='dark:red', facecolor=[0,0,0,0], edgecolor='blue', linewidth=5)
plt.show() 
```

# pairplot : combination of scatter plot and histogram



```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.pairplot(data =df)
plt.show()
```

```python
sns.pairplot(data= df, height=1.5, hue='species')
plt.show()
```

```python
sns.pairplot(data= df, height=1.5, hue='species', diag_kind='hist')
plt.show()
```

```python
sns.pairplot(data= df, height=1.5, hue='species', diag_kind='kde', plot_kws={'edgecolor':'white'})
plt.show()
```

# catplot : it is categorial plot means you can make any plot using these and by default it work on scatter plot

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='sex', y='tip', data=df, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='tip', data=df, hue='smoker', col='time', height=3)
plt.show()
```

```python
ds= sns.load_dataset('titanic')
ds.head()
```

```python
sns.catplot(x='class', data=ds, col='deck', col_wrap=3, height=3)
plt.show()
```

# strip plot : same as scatter plot but have different functionality

```python
df= sns.load_dataset('tips')
df.head()
```

```python
sns.stripplot(x='day', y='tip', data=df, hue='sex', jitter=0.9)
plt.show()
# jitter is used to seen data point clearly
```

```python
sns.stripplot(x='day', y='tip', data=df, hue='sex', dodge=True)
plt.show()
# dodge give seperate dots for the same which are not in conjested manner
```
