# strip plot : same as scatter plot but have different functionality

```python
df= sns.load_dataset('tips')
df.head()
```

```python
sns.stripplot(x='day', y='tip', data=df, hue='sex', jitter=0.9)
plt.show()
# jitter is used to seen data point clearly
```

```python
sns.stripplot(x='day', y='tip', data=df, hue='sex', dodge=True)
plt.show()
# dodge give seperate dots for the same, which are not in conjested manner
```

# multiport(facetgrid) : adding multiple plot in once

```python
df = sns.load_dataset('exercise')
df.head()
```

```python
mp = sns.FacetGrid(data=df, col='kind', hue='time')
mp.map(sns.scatterplot, 'time', 'pulse')
mp.add_legend()
plt.show()
```

```python
mp = sns.FacetGrid(data=df, col='kind', hue='time')
mp.map(sns.barplot, 'time', 'pulse', order=['30 min','1 min', '15 min'])
mp.add_legend()
plt.show()
```

# swarm plot : more like strip plot but it's more readable and won't overlap

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.swarmplot(x='day', y='tip', data=df, hue='sex', dodge=True)
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=df, hue='sex', dodge=True)
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=df, hue='sex', dodge=True, palette='dark:yellow', order=['Sun','Thur','Fri','Sat'], marker='^', alpha=0.9, linewidth=0.5, edgecolor='black')
plt.show()
```
